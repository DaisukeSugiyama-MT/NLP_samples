```python
a = np.array([1,3,4])
b = np.array([3,4,5])

distance = np.linalg.norm(b-a)
print(distance)
# 3次元の2点の距離
# https://www.higashisalary.com/entry/numpy-linalg-norm
```

```python
a = np.array([1,3,4])
b = np.array([3,4,5])
vec = b - a

distance = np.arctan2(vec[0],vec[1])
print(distance)
# 3次元の2点の角度
# https://kagglenote.com/ml-tips/between-two-points/
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def cos_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

x = np.array([6,4,11])
y = np.array([9,14,17])
z = np.array([207,242,255])

print(cos_similarity(x, y))
print(cos_similarity(x, z))
print(cos_similarity(y, z))

x = [[0,0,255]]
y = [[0,255,255]]
z = [[255,255,0]]

print(cosine_similarity(x,y))
print(cosine_similarity(x,z))
print(cosine_similarity(y,z))
```
